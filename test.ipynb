```python
import numpy as np

# 假设ρ1和ρ2是你的电价表
ρ1 = np.array([3.5, 2.8, 2.2, 3.0])  # 代表[0,30)时刻的电价
ρ2 = np.array([1.8, 2.1, 2.5, 2.0])  # 代表[30,96)时刻的电价

# 合并ρ1和ρ2成ρ
ρ = np.concatenate((ρ1, ρ2), axis=None)

# 获取ρ1在ρ中的索引位置
t1 = np.where(np.isin(ρ, ρ1))

# 获取ρ2在ρ中的索引位置
t2 = np.where(np.isin(ρ, ρ2))

# t1包含ρ1在ρ中的索引位置，t2包含ρ2在ρ中的索引位置
print(type(t1))
```

```python
import numpy as np

# 假设你有一个包含元组的列表，每个元组包含一个numpy矩阵和其适应度值
data = [
    (np.array([[1, 2, 3], [4, 5, 6]]), 10),
    (np.array([[7, 8, 9], [10, 11, 12]]), 5),
    (np.array([[13, 14, 15], [16, 17, 18]]), 7)
]

# 使用sorted函数按适应度值从小到大排序列表，并得到排序后的索引
sorted_indices = sorted(range(len(data)), key=lambda i: data[i][1])

# 打印排序后的索引
print(sorted_indices)
```
